# PLOT PYTHON

 ## Line PLOT

```python
import numpy as np
from matplotlib import pyplot as plt

t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

plt.figure()
plt.plot(t, s)
plt.title('Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (mV)')
plt.grid()
plt.show()
```
